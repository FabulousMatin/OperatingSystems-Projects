## Execution times

```python

import matplotlib.pyplot as plt

data = []
with open("execution_times.txt", "r") as file:
    for line in file:
        nthreads, duration = line.strip().split()
        data.append((int(nthreads), int(duration)))

nthreads = [entry[0] for entry in data]
duration = [entry[1] for entry in data]


plt.figure(figsize=(15,6))
plt.plot(nthreads, duration, 'bo-')
plt.vlines(nthreads, 0, duration, colors='r', lw=2, alpha=0.5)
plt.xticks(range(0,len(duration) + 1,2))
plt.xlabel('threads')
plt.ylabel('duration')
plt.title('Execution Times')
plt.show()
```

## Speed Up

$$ speedup = \frac{\text{serial execution time}}{\text{paralle execution time}} $$

```python
speedup = [duration[0] / x for x in duration]
```

```python
plt.figure(figsize=(15,6))
plt.plot(nthreads, speedup, 'bo-')
plt.vlines(nthreads, 0, speedup, colors='r', lw=2, alpha=0.5)
plt.xticks(range(0,len(speedup) + 1,2))
plt.xlabel('threads')
plt.ylabel('speed up')
plt.title('Execution Speed-up')
plt.show()
```

- you can see the highest speedup ratio, happens in number of threads = 7


- increasing number of threads will not always result in increasing ratio(~ decreasing exec time), beacause making/handling/syncing threads adds some overhead.
